# The effects of the Covid 19 in Argentina using Apache Spark: a comparative analisys between provinces.

We download "https://sisa.msal.gov.ar/datos/descargas/covid-19/files/Covid19Casos.csv" from "datos.gob.ar" using Pandas library.
Then we create a dataframe using Pyspark and we filter it taking into account only confirmed cases. Finally we divided it between those who recovered and those who passed away, showing it in two graphs.

*Beware: the Covid19Casos.csv file is 1,2gb it may take a while to download it.*

```python
import pandas as pd
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
from pyspark.sql.functions import col
import matplotlib.pyplot as plt
import numpy as np

pdDF = pd.read_csv('https://sisa.msal.gov.ar/datos/descargas/covid-19/files/Covid19Casos.csv')

sc = SparkContext.getOrCreate()
spark = SparkSession(sc)

df = spark.createDataFrame(pdDF)
```

```python
casos_confirmados = df.select('edad', 'residencia_provincia_nombre','fallecido').where(col('clasificacion_resumen').like('Confirmado'))

Tierra_del_Fuego = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Tierra del Fuego')).count()
San_Luis = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('San Luis')).count()
Chubut = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Chubut')).count()
Jujuy = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Jujuy')).count()
Neuquén = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Neuquén')).count()
CABA = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('CABA')).count()
Santa_Cruz = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Santa Cruz')).count()
Entre_Ríos = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Entre Ríos')).count()
Tucumán = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Tucumán')).count()
La_Pampa = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('La Pampa')).count()
Chaco = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Chaco')).count()
Mendoza = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Mendoza')).count()
Santa_Fe = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Santa Fe')).count()
Corrientes = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Corrientes')).count()
Buenos_Aires = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Buenos Aires')).count()
Formosa = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Formosa')).count()
Córdoba = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Córdoba')).count()
Salta = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Salta')).count()
San_Juan = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('San Juan')).count()
Santiago_del_Estero = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Santiago del Estero')).count()
Misiones = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Misiones')).count()
Catamarca = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Catamarca')).count()
Río_Negro = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('Río Negro')).count()
La_Rioja = casos_confirmados.select('residencia_provincia_nombre').where(col('residencia_provincia_nombre').like('La Rioja')).count()
```

```python
labels = ['Buenos Aires','CABA','Catamarca','Chaco','Chubut','Córdoba','Corrientes','Entre Ríos','Formosa','Jujuy',
          'La Pampa','La Rioja','Mendoza','Misiones','Neuquén','Río Negro','Salta','San Juan','San Luis',
          'Santa Cruz','Santa_Fe','Santiago del Estero','Tierra del Fuego','Tucumán']

values = [Buenos_Aires,CABA,Catamarca,Chaco,Chubut,Córdoba,Corrientes,Entre_Ríos,Formosa,Jujuy,La_Pampa,
          La_Rioja,Mendoza,Misiones,Neuquén,Río_Negro,Salta,San_Juan,San_Luis,Santa_Cruz,Santa_Fe,
          Santiago_del_Estero,Tierra_del_Fuego,Tucumán]

y = np.arange(100000,900000, 200000)

plt.bar(labels, values)
plt.gcf().set_size_inches(30, 15)
plt.title('Total de contagiados por provincia', fontdict= {'fontsize':40})
plt.xlabel('Provincias',fontdict= {'fontsize':30})
plt.ylabel('Cantidad de contagiados',fontdict= {'fontsize':30})
plt.yticks(y)
plt.grid()
plt.show()
```
